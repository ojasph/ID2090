```python
import numpy as np
import matplotlib.pyplot as plt

N= 1000
radius = int(input('Enter the radius value, which is an integer and less than or equal to 50:  '))

circleX, circleY = [], []
for i in range(360):
    circleX.append(50 + radius*np.cos(np.pi*i/180))
    circleY.append(50 + radius*np.sin(np.pi*i/180))

xdata = 100*np.random.random(N)
ydata = 100*np.random.random(N)

def detective(x1,y1,ins,outs):
    if (x1 - 50)**2 + (y1 - 50)**2 <= radius**2:
        ins=1
        return ins
    else: 
        return outs

plt.figure(figsize=(7,7))
  
inside, outside = 0, 0
for i in range(N):
    if detective(xdata[i], ydata[i], 0, 0) == 1:
        inside +=1
    else:
        outside +=1
print('Ratio of points inside vs outside is: ', inside/outside)

for i in range(N):
    if detective(xdata[i], ydata[i], 0, 0) == 1:
        plt.scatter(xdata[i], ydata[i], c = 'b', s = 3)
    else:
        plt.scatter(xdata[i], ydata[i], c = 'g', s = 1.5)

plt.plot(circleX, circleY, color = 'r')
plt.xlim(0,100)
plt.ylim(0,100)
print('Value of pi is: ', (100/radius)**2*inside/N)
```

```python
value = []
xmax= 1000
for i in range(xmax):   
    N= 1000
    radius = 40
    
    circleX, circleY = [], []
    for i in range(361):
        circleX.append(50 + radius*np.cos(np.pi*i/180))
        circleY.append(50 + radius*np.sin(np.pi*i/180))
  
    xdata = 100*np.random.random(N)
    ydata = 100*np.random.random(N)

    def detective(x1,y1,inside):
        if (x1 - 50)**2 + (y1 - 50)**2 <= radius**2:
            inside=1
        return inside
    
    p = []
    for i in range(N):
        p.append(detective(xdata[i], ydata[i], 0))
    
    plt.xlim(0,xmax-1)
    plt.ylim(2.7,3.6)
    
    c= sum(p)
    
    value.append((100/radius)**2*c/N) 

plt.plot(value)
value.sort()
plt.plot(value)
plt.show()

m= int(xmax/2)
print('The median is:' , value[m])
print('The mean is:' , sum(value)/xmax)
print('The range is :', [min(value), max(value)])
print('Standard deviation is', np.std(value))
```
