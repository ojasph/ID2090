```python
u = [5, 2, 3]
v = [10, 15, 9]
arr_new = [[None for _ in range(3)] for _ in range(3)]

# Define the calculate_distance() function
def calculate_distance(x1, y1, x2, y2):
    # Replace this with your actual distance calculation logic
    return ((x2 - x1)**2 + (y2 - y1)**2)**0.5

# Calculate the distance between each pair of points and store in arr_new
for i in range(3):
    for j in range(3):
        arr_new[i][j] = ([u[i], v[i]], [u[j], v[j]], calculate_distance(u[i], v[i], u[j], v[j]))

# Sort each row of arr_new based on the calculated distance
for row in arr_new:
    row.sort(key=lambda tup: tup[2])

print(arr_new)
```
