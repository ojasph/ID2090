```python
var('x')
f(x) = x^5 -13*x^3 - x^2 + 10*x + 170 
solution = solve(f, x, to_poly_solve=True)
print(solution)
```

```python
A = Matrix([[0.1,0.1,0.2, 0.1, 0],
            [0,0.55,0, 0.6, 0],
            [0,0,0.3,2,0], 
            [0,0,0,0.33,0], 
            [0,0.3,0,0,4]])
A.is_invertible()
B = A.inverse()
D = Matrix([[0,0,0,0,-170],
            [1,0,0,0,-10],         
            [0,1,0,0,1],         
            [0,0,1,0,13],         
            [0,0,0,1,0]])
E = A*D*B
```

```python
plot(f(x), x, axes_labels=['x', 'f(x)'])
```

```python
Q = []
for i in range(5):
    Q.append(solution[i].rhs())
```

```python
def complex_point_plot(pts): 
    return list_plot([(real(i), imag(i)) for i in pts], axes_labels = ['Re($z$)', 'Im($z$)'], size=30)
points_involved = complex_point_plot(Q)
b = plot(f(x), x)

plot(points_involved)
```

```python
radii = []
centre_list = []
def radius(P):
    i = -1
    for row in P:
        i+=1
        sum = 0
        for element in row:
            element = abs(element)
            sum +=element   
        sum = sum-abs(row[i])
        radii.append(sum)
        centre_list.append(row[i])
radius(E)
p=[]
for i in range(5):
    c = circle((centre_list[i],0), radii[i], rgbcolor=(1,0,0))
    p.append(plot(c))
a = sum(p)

def complex_point_plot(pts): 
    return list_plot([(real(i), imag(i)) for i in pts], axes_labels = ['Re($z$)', 'Im($z$)'], size=30)
b = complex_point_plot(Q)

c = (b+ a)
c.plot()
```

```python
def dist(Z):
    return norm(vector(Z))
dis = []
for i in range(5):
    dis.append(sqrt(norm(Q[i])))

sorted(dis, key=None, reverse=False)
dis.sort(reverse = True)
print("Radius of the Circle is" + str(dis[0]))
circle_norm = circle((0,0), dis[0])
d = (b+circle_norm)
d.plot()
```
